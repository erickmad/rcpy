# Reservoirpy

## Creating a reservoir

```python
import reservoirpy as rpy
from reservoirpy.nodes import Reservoir
from reservoirpy.mat_gen import uniform
import numpy as np

# Create a reservoir
rpy.set_seed(42)

reservoir_units = 100
spectral_radius = 0.9
input_scaling = 8
average_degree = 10
rc_connectivity = average_degree/reservoir_units

res = Reservoir(units=reservoir_units, Win=uniform(low=-1.0, high=1.0), W=uniform(low=-1.0, high=1.0), sr=0.9,input_scaling=input_scaling, input_connectivity=0.9, rc_connectivity=rc_connectivity)

# Run once with some dummy data to *initialize* weights
X = np.random.rand(10, 1)       # e.g., 10 timesteps, 3 input features
_ = res.run(X)

# Access the matrices
Win_matrix = res.Win            # shape: (units, input_dim)
W_matrix   = res.W              # shape: (units, units)

print("Input weights (Win):", Win_matrix.shape)
print("Reservoir weights (W):",   W_matrix.shape)
```

```python
print("Input matrix (Win):")
print(res.Win.toarray())

print("\nReservoir matrix (W):")
print(res.W.toarray())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("Input Weights (Win)")
plt.imshow(res.Win.toarray(), aspect='auto', cmap='viridis')
plt.colorbar()
plt.subplot(1, 2, 2)
plt.title("Reservoir Weights (W)")
plt.imshow(res.W.toarray(), aspect='auto', cmap='viridis')
plt.colorbar()
plt.tight_layout()
plt.show()
```
