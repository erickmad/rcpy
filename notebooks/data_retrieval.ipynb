## Import ENSO data

Data retrieved from NOAA's Physical Sciences Laboratory Monthly Climate/Ocean Indices webpage [1].
- Niño 3.4 index (1870-2024): ([HadISST1.1](https://psl.noaa.gov/data/timeseries/month/DS/Nino34/))
- Niño 3.4 index (1950-2024): ([ERSST V5](https://psl.noaa.gov/data/timeseries/month/DS/Nino34_CPC/))

### Quick visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from rcpy.data import ClimateIndex, load_NOAA_data

data = load_NOAA_data(ClimateIndex.ONI)

index_values = data['index']
index_values = index_values[:-12]
# Create monthly date range from the start and end years
start_date = f"1950-01-01"
end_date = f"2024-12-31"
date_range = pd.date_range(start=start_date, end=end_date, freq='ME')
# Create DataFrame with date range and index values

# Plot values and anomalies
plt.figure(figsize=(10, 5))
plt.plot(date_range, index_values, label='nino', color='tab:blue')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title(f'{data["full_name"]} anomalies')
plt.tight_layout()

# Show the plot
plt.show()
```
